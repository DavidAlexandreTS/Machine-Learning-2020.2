```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
# Make a dataset with the csv file
# None indicate there first line would be a line, not columns
store_data = pd.read_csv('/content/Market_Basket_Optimisation.csv', header = None)
```

```python
# Show the format of the dataset
store_data.head()
```

```python
# List to get all lines of the dataset
data_set_lines = []
```

```python
# Read all lines of the dataset and put it on data_set_lines
for i in range(0, 7501):
  data_set_lines.append([str(store_data.values[i, j]) for j in range(0, 20)])
```

```python
# Shows the list of lists dataset lines where each list are a line from dataset
print(data_set_lines)
```

```python
# Generate Association Rules with the Apriori Algorithm
# this apriori needs some values to work: the first is the list of list that we go extract rules
# the second is the minimum support, the third is the minimum confidence
# followed by the minimum lift and the minimum legth(number minimum of items in the rule) respectively
association_rules = apriori(data_set_lines, min_support = 0.0045, min_confidence = 0.2, min_lift = 3, min_length = 2)
# Transform association_rules in a list
association_results = list(association_rules)
```

 ***Note: Our dataset has one week purchases, but we will consider a minimum support only products that are purchased at least 5 time in one day, so we have 5 x 7 = 35 and 35/7500 = 0.0045. Futhermore, on this implementation, we have the min_lift parameter, this parameter specifies the minimum lift value for the short listed rules, in other words, tells us that a product x is minimum n times more likely to be bought by the customers who buy a product y compared to the default likelihood of the sale of x.***

```python
# Shows the amount of rules found
print("Were found", len(association_results), "association rules")
```

```python
# Visualize the rules
for i in range(len(association_results)):
  print(association_results[i])
  print()
```
